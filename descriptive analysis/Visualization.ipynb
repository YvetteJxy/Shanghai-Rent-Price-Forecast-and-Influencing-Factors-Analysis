```python
# （1）房屋类型与租金的关系
# 属于房屋类型的数据有 房屋面积、客厅数量、厕所数量、卧室数量、楼层相对位置、房屋类型、是否朝南
# 1.1 房屋租金与房屋类型的关系
data.head()
```

```python
# 增加一列字符型数据表示房屋类型，便于更直观得可视化
def transform_string(df):
    if df["房屋类型"] == 2:
        return "复式"
    elif df["房屋类型"] == 3:
        return "跃层"
    else:
        return "平房"


data.loc[:,"房屋类型（字符串）"] = data.apply(transform_string,axis = 1)
data.head()
```

```python
data["房屋类型"].unique()
```

```python
# 增加一列字符型数据表示房屋类型，便于更直观得可视化
import math
for i,item in enumerate(data["每月租金"]):
    data.loc[i, "log(每月租金)"] = math.log(item)
data.head()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x="房屋类型（字符串）", y="每月租金", data=data)
sns.set_palette("husl")
```

```python
# 由箱线图可知，三类房屋类型中，复式的租金普遍高于跃层，跃层租金又略高于平房、
# 按照常识来看跃层要比复式房屋更为高级，这样的数据分布显然与常识不符
# 通过回到网页，观测原始数据发现，链家网的关于复式和跃层的规格区分不明显，有的严格上属于loft的房屋被分到跃层，而真正的跃层房屋又被分到loft
# 因此，我们直接将房屋类型的数据重新进行分类，平房部分不变，将复式和跃层统一归为“多层”
data.head()
```

```python
# 将房屋类型中2、3两类统一为2
for i,item in enumerate(data["房屋类型"]):
    if data["房屋类型"][i] == 3:
        data["房屋类型"][i] = 2
```

```python
data["房屋类型"].unique()
```

```python
data.drop(labels = "房屋类型（字符串）" , axis = 1, inplace = True)
data.head()
```

```python
# 增加一列字符型数据表示房屋朝向，便于更直观得可视化
def transform_string3(df):
    if df["是否朝南"] == 1:
        return "朝南"
    else:
        return "不朝南"


data.loc[:,"房屋朝向"] = data.apply(transform_string3,axis = 1)
data.head()
```

```python
# 根据图片信息，猜测月租对房屋朝向信息不敏感，进行显著性检验
# Kruskal-Wallis检验

# Kruskal-Wallis H Test
from scipy.stats import kruskal
```

```python
data.head()
```

```python
data["是否朝南"]==1
```

```python
data1 = data.loc[data["是否朝南"]==1,"每月租金"]
```

```python
data2 = data.loc[data["是否朝南"]==0,"每月租金"]
```

```python
stat, p = kruskal(data1, data2)
print("stat={},p={}".format(stat,p))
if p > 0.05:
    print('不能拒绝原假设，样本集分布相同')
else:
    print('拒绝原假设，样本集分布可能不同')
```
